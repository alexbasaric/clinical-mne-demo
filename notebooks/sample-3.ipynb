```python
import urllib.request

url = "https://physionet.org/files/eegmmidb/1.0.0/S001/S001R01.edf"
file_name = "sample_eeg.edf"

urllib.request.urlretrieve(url, file_name)

print("Downloaded:", file_name)
```

```python
import mne
raw = mne.io.read_raw_edf("sample_eeg.edf", preload=True)
print(raw)
```

```python
raw.plot(duration=10, n_channels=10)
```

```python
print(raw.info)
```

```python
raw_filt = raw.copy().filter(1, 30)
raw_filt.plot(duration=10, n_channels=10)
```
