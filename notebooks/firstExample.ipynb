```python
import numpy as np
import mne

sfreq = 100  # Hz
times = np.arange(1000) / sfreq

amplitude_uv = 20  # 20 mikrovolti
data = (amplitude_uv * 1e-6 * np.sin(2 * np.pi * 10 * times))[None, :]

info = mne.create_info(ch_names=["EEG 001"], sfreq=sfreq, ch_types=["eeg"])
raw = mne.io.RawArray(data, info)

print(raw)
raw.plot()
```

```python
import numpy as np
import mne

sfreq = 200
times = np.arange(2000) / sfreq  # 10 sekundi

sig1 = 20e-6 * np.sin(2 * np.pi * 10 * times)   # 10 Hz, 20 µV
sig2 = 10e-6 * np.sin(2 * np.pi * 3 * times)    # 3 Hz, 10 µV

data = (sig1 + sig2)[None, :]

info = mne.create_info(["EEG 001"], sfreq=sfreq, ch_types=["eeg"])
raw = mne.io.RawArray(data, info)

raw.plot(scalings={"eeg": 50e-6}, duration=5)
```
